```python
# basic info about items, we're not gonna go into the deep level of analyzing nutrion, for example it's not necessary to know which amount of sodium a burger gives us!!
b_data = data.drop(columns=['Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                         'Iron (% Daily Value)', 'Calcium (% Daily Value)', 'Dietary Fiber (% Daily Value)', 'Dietary Fiber', 'Sodium (% Daily Value)', 'Sodium', 'Cholesterol (% Daily Value)', 'Saturated Fat',
                         'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol', 'Calories from Fat', 'Total Fat (% Daily Value)', 'Carbohydrates (% Daily Value)'])
b_data
```

```python
categories = b_data['Category'].unique()
categories
```

```python
# mean of calorie for each category
avg_calories = pd.DataFrame({'Category': b_data['Category'].unique()}, columns=[
                            'Category', 'avg_calories'])

for item in categories:
    avg_calories['avg_calories'][avg_calories['Category'] ==
                                 item] = b_data[b_data['Category'] == item]['Calories'].mean()

avg_calories
```
