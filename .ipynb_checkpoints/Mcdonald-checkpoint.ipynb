```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv('menu.csv')
data
```

```python
# checking is there is NaN value or not
data.isnull().sum().sum()
```

```python
data.columns
```

```python
# basic info about items, we're not gonna go into the deep level of analyzing nutrion, for example it's not necessary to know which amount of sodium a burger gives us!!
b_data = data.drop(columns=['Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
                         'Iron (% Daily Value)', 'Calcium (% Daily Value)', 'Dietary Fiber (% Daily Value)', 'Dietary Fiber', 'Sodium (% Daily Value)', 'Sodium', 'Cholesterol (% Daily Value)', 'Saturated Fat',
                         'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol', 'Calories from Fat', 'Total Fat (% Daily Value)', 'Carbohydrates (% Daily Value)'])
b_data
```

```python
categories = b_data['Category'].unique()
categories
```

```python
# mean of calorie for each category
avg_calories = pd.DataFrame({'Category': b_data['Category'].unique()}, columns=[
                            'Category', 'avg_calories', 'Carbs'])

for item in categories:
    avg_calories['avg_calories'][avg_calories['Category'] ==
                                 item] = b_data[b_data['Category'] == item]['Calories'].mean()
    avg_calories['Carbs'][avg_calories['Category']
                          == item] = b_data[b_data['Category'] == item]['Carbohydrates'].mean()

avg_calories
```

```python
plt.figure(figsize=(12, 6))
barplot = sns.barplot(
    x=avg_calories['Category'], y=avg_calories['avg_calories'])
#plt.setp(barplot.get_xticklabels(), rotation=90)
```

```python
colors = ['aqua', 'darkblue', 'darkgreen', 'red',
          'plum', 'orange', 'teal', 'maroon', 'yellowgreen']
avg_calories['Carbs'] = avg_calories['Carbs'].astype(int)
plt.figure(figsize=(15, 10))
plot = plt.scatter(x=avg_calories['Category'], y=avg_calories['avg_calories'], s=(
    avg_calories['Carbs']*50), c=colors)
plt.xlabel('Category', fontsize=16)
plt.ylabel('Calories', fontsize=16)
plt.show()
```

```python
drinks = b_data[b_data['Category'] == 'Coffee & Tea'].sort_values(by='Calories', ascending=False)
drinks
```

```python
drinks = drinks.assign(Label = pd.DataFrame(['NaN']))
for ind in drinks.index:
    if drinks['Calories'][ind] >= 500:
        drinks['Label'][ind] = 'Super High In Calories'
    elif drinks['Calories'][ind] < 500 and drinks['Calories'][ind] >= 300:
        drinks['Label'][ind] = 'High In Calories'
    elif drinks['Calories'][ind] < 300 and drinks['Calories'][ind] >= 150:
        drinks['Label'][ind] = 'Not So High, Not So Low'
    elif drinks['Calories'][ind] < 150 and drinks['Calories'][ind] > 0:
        drinks['Label'][ind] = 'Low In Calories'
    else:
        drinks['Label'][ind] = 'Zero Calorie'
drinks
        
        
    
```

```python
labels = drinks['Label'].unique()
colors = colors = ['#787da5', '#e07a5f', '#3d405b', '#81b29a','#f2cc8f' ]
sizes = dict.fromkeys(labels, 0)
for item in sizes.keys():
    for loc in drinks.index:
        if drinks['Label'][loc] == item:
            sizes[item]+=1

for item in sizes.keys():
    sizes[item] = int((sizes[item]/85)*100)

sizes.values()
    
fig, ax = plt.subplots()
ax.pie(sizes.values(), colors=colors, labels=labels, autopct='%1.1f%%')
centre_circle = plt.Circle((0,0),0.75,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax.axis('equal')  
plt.tight_layout()

plt.show()
```
